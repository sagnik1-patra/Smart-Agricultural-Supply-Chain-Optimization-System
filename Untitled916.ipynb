```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from math import gamma, sin, pi

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.cluster import KMeans

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ============================================================
# Paths
# ============================================================

INPUT_FILE = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Supply Chain Optimization System\megambmarketname_n.xls"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Supply Chain Optimization System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "csa_"

# ============================================================
# Load Dataset
# ============================================================

df = pd.read_excel(INPUT_FILE)

df.columns = (
    df.columns.astype(str)
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

df = df.drop_duplicates()
df = df.fillna("Unknown")

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("Columns:", list(df.columns))

# ============================================================
# Detect Market and District Columns
# ============================================================

market_col = None
district_col = None

for col in df.columns:
    if "market" in str(col).lower():
        market_col = col

    if "district" in str(col).lower():
        district_col = col

if market_col is None:
    market_col = df.columns[0]

if district_col is None:
    if len(df.columns) > 1:
        district_col = df.columns[1]
    else:
        df["District"] = "Unknown"
        district_col = "District"

print("Market Column:", market_col)
print("District Column:", district_col)

# ============================================================
# Synthetic Supply Chain Features
# ============================================================

np.random.seed(42)

df["Market_ID"] = range(1, len(df) + 1)
df["Latitude"] = np.random.uniform(25.0, 26.8, len(df))
df["Longitude"] = np.random.uniform(89.5, 92.8, len(df))

df["Demand_Tons"] = np.random.randint(50, 500, len(df))
df["Supply_Tons"] = np.random.randint(40, 550, len(df))
df["Storage_Capacity_Tons"] = np.random.randint(100, 1000, len(df))
df["Transport_Cost_Per_KM"] = np.random.uniform(15, 60, len(df))
df["Road_Quality_Index"] = np.random.uniform(0.4, 1.0, len(df))
df["Seasonal_Risk_Index"] = np.random.uniform(0.1, 0.9, len(df))

df["Demand_Supply_Gap"] = df["Demand_Tons"] - df["Supply_Tons"]
df["Market_Load_Index"] = df["Demand_Tons"] / (df["Storage_Capacity_Tons"] + 1)

df["Estimated_Logistics_Cost"] = (
    abs(df["Demand_Supply_Gap"]) * df["Transport_Cost_Per_KM"]
    + df["Seasonal_Risk_Index"] * 1000
    - df["Road_Quality_Index"] * 300
)

df["Estimated_Logistics_Cost"] = df["Estimated_Logistics_Cost"].clip(lower=100)

# ============================================================
# Encoding
# ============================================================

label_encoders = {}

for col in [market_col, district_col]:
    le = LabelEncoder()
    df[col + "_Encoded"] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

# ============================================================
# Features and Target
# ============================================================

feature_cols = [
    market_col + "_Encoded",
    district_col + "_Encoded",
    "Latitude",
    "Longitude",
    "Demand_Tons",
    "Supply_Tons",
    "Storage_Capacity_Tons",
    "Transport_Cost_Per_KM",
    "Road_Quality_Index",
    "Seasonal_Risk_Index",
    "Demand_Supply_Gap",
    "Market_Load_Index"
]

target_col = "Estimated_Logistics_Cost"

X = df[feature_cols]
y = df[target_col]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================
# CSA Feature Selection Functions
# ============================================================

def evaluate_feature_subset(mask, X_data, y_data):
    if np.sum(mask) == 0:
        return 999999

    selected_X = X_data[:, mask == 1]

    X_train_fs, X_test_fs, y_train_fs, y_test_fs = train_test_split(
        selected_X,
        y_data,
        test_size=0.2,
        random_state=42
    )

    model = RandomForestRegressor(
        n_estimators=100,
        random_state=42
    )

    model.fit(X_train_fs, y_train_fs)
    pred = model.predict(X_test_fs)

    rmse = np.sqrt(mean_squared_error(y_test_fs, pred))
    feature_penalty = 0.01 * np.sum(mask)

    return rmse + feature_penalty


def levy_flight(beta=1.5, size=1):
    sigma = (
        gamma(1 + beta)
        * sin(pi * beta / 2)
        /
        (
            gamma((1 + beta) / 2)
            * beta
            * (2 ** ((beta - 1) / 2))
        )
    ) ** (1 / beta)

    u = np.random.normal(0, sigma, size)
    v = np.random.normal(0, 1, size)

    step = u / (np.abs(v) ** (1 / beta))

    return step


def sigmoid(x):
    x = np.clip(x, -10, 10)
    return 1 / (1 + np.exp(-x))


def create_valid_mask(n_features):
    mask = np.random.randint(0, 2, size=n_features)

    if np.sum(mask) == 0:
        mask[np.random.randint(0, n_features)] = 1

    return mask


def csa_feature_selection(
    X_data,
    y_data,
    n_nests=20,
    generations=30,
    discovery_rate=0.25
):
    n_features = X_data.shape[1]

    nests = np.array([
        create_valid_mask(n_features)
        for _ in range(n_nests)
    ])

    fitness = np.array([
        evaluate_feature_subset(nest, X_data, y_data)
        for nest in nests
    ])

    best_index = np.argmin(fitness)
    best_nest = nests[best_index].copy()
    best_fitness = fitness[best_index]

    fitness_history = []

    for generation in range(generations):

        for i in range(n_nests):
            step = levy_flight(size=n_features)
            probability = sigmoid(step)

            new_nest = nests[i].copy()

            mutation_mask = np.random.rand(n_features) < probability
            new_nest[mutation_mask] = 1 - new_nest[mutation_mask]

            if np.sum(new_nest) == 0:
                new_nest[np.random.randint(0, n_features)] = 1

            new_fitness = evaluate_feature_subset(new_nest, X_data, y_data)

            random_index = np.random.randint(0, n_nests)

            if new_fitness < fitness[random_index]:
                nests[random_index] = new_nest
                fitness[random_index] = new_fitness

        abandon_count = max(1, int(discovery_rate * n_nests))
        worst_indices = np.argsort(fitness)[-abandon_count:]

        for idx in worst_indices:
            nests[idx] = create_valid_mask(n_features)
            fitness[idx] = evaluate_feature_subset(nests[idx], X_data, y_data)

        current_best_index = np.argmin(fitness)

        if fitness[current_best_index] < best_fitness:
            best_fitness = fitness[current_best_index]
            best_nest = nests[current_best_index].copy()

        fitness_history.append(best_fitness)

        print(
            f"CSA Generation {generation + 1}/{generations} | "
            f"Best Fitness: {best_fitness:.4f}"
        )

    return best_nest, best_fitness, fitness_history


# ============================================================
# Run CSA
# ============================================================

best_mask, best_fitness, fitness_history = csa_feature_selection(
    X_scaled,
    y.values,
    n_nests=20,
    generations=30,
    discovery_rate=0.25
)

selected_features = [
    feature for feature, selected in zip(feature_cols, best_mask)
    if selected == 1
]

print("\nCSA Selected Features:")
print(selected_features)

X_selected = X_scaled[:, best_mask == 1]

# ============================================================
# Train Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_selected,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================================
# CSA Random Forest Model
# ============================================================

csa_rf_model = RandomForestRegressor(
    n_estimators=300,
    max_depth=12,
    random_state=42
)

csa_rf_model.fit(X_train, y_train)

rf_pred = csa_rf_model.predict(X_test)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_r2 = r2_score(y_test, rf_pred)
rf_accuracy = max(0.0, float(rf_r2)) * 100

# ============================================================
# CSA Neural Network Model
# ============================================================

csa_nn_model = Sequential([
    Input(shape=(X_train.shape[1],)),
    Dense(128, activation="relu"),
    Dropout(0.2),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1)
])

csa_nn_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = csa_nn_model.fit(
    X_train,
    y_train,
    validation_split=0.1,
    epochs=200,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

nn_pred = csa_nn_model.predict(X_test).flatten()

nn_mae = mean_absolute_error(y_test, nn_pred)
nn_rmse = np.sqrt(mean_squared_error(y_test, nn_pred))
nn_r2 = r2_score(y_test, nn_pred)
nn_accuracy = max(0.0, float(nn_r2)) * 100

# ============================================================
# Results
# ============================================================

results_df = pd.DataFrame([
    {
        "Model": "CSA Random Forest",
        "MAE": rf_mae,
        "RMSE": rf_rmse,
        "R2_Score": rf_r2,
        "Accuracy_Percentage": rf_accuracy,
        "Selected_Features": len(selected_features)
    },
    {
        "Model": "CSA Neural Network",
        "MAE": nn_mae,
        "RMSE": nn_rmse,
        "R2_Score": nn_r2,
        "Accuracy_Percentage": nn_accuracy,
        "Selected_Features": len(selected_features)
    }
])

best_model_name = results_df.sort_values(
    by="R2_Score",
    ascending=False
).iloc[0]["Model"]

print("\nBest Model:", best_model_name)
print(results_df)

# ============================================================
# Full Dataset Prediction
# ============================================================

if best_model_name == "CSA Neural Network":
    full_predictions = csa_nn_model.predict(X_selected).flatten()
else:
    full_predictions = csa_rf_model.predict(X_selected)

df["CSA_Predicted_Logistics_Cost"] = full_predictions

df["CSA_Market_Rank"] = df["CSA_Predicted_Logistics_Cost"].rank(
    ascending=False,
    method="dense"
).astype(int)

# ============================================================
# KMeans Clustering
# ============================================================

cluster_features = df[
    [
        "Demand_Tons",
        "Supply_Tons",
        "Storage_Capacity_Tons",
        "Transport_Cost_Per_KM",
        "Road_Quality_Index",
        "Seasonal_Risk_Index"
    ]
]

cluster_scaler = StandardScaler()
cluster_scaled = cluster_scaler.fit_transform(cluster_features)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["CSA_Supply_Chain_Cluster"] = kmeans.fit_predict(cluster_scaled)

cluster_labels = {
    0: "Low Priority Market",
    1: "Medium Priority Market",
    2: "High Priority Market"
}

df["CSA_Cluster_Label"] = df["CSA_Supply_Chain_Cluster"].map(cluster_labels)

# ============================================================
# Save CSV Files
# ============================================================

processed_csv = os.path.join(OUTPUT_DIR, PREFIX + "processed_dataset.csv")
result_csv = os.path.join(OUTPUT_DIR, PREFIX + "model_results.csv")
prediction_csv = os.path.join(OUTPUT_DIR, PREFIX + "prediction_results.csv")
selected_features_csv = os.path.join(OUTPUT_DIR, PREFIX + "selected_features.csv")
cluster_csv = os.path.join(OUTPUT_DIR, PREFIX + "cluster_results.csv")
fitness_csv = os.path.join(OUTPUT_DIR, PREFIX + "fitness_history.csv")

df.to_csv(processed_csv, index=False)
results_df.to_csv(result_csv, index=False)

df[
    [
        market_col,
        district_col,
        "Demand_Tons",
        "Supply_Tons",
        "Estimated_Logistics_Cost",
        "CSA_Predicted_Logistics_Cost",
        "CSA_Market_Rank",
        "CSA_Cluster_Label"
    ]
].to_csv(prediction_csv, index=False)

pd.DataFrame({
    "Selected_Feature": selected_features
}).to_csv(selected_features_csv, index=False)

df[
    [
        market_col,
        district_col,
        "CSA_Supply_Chain_Cluster",
        "CSA_Cluster_Label"
    ]
].to_csv(cluster_csv, index=False)

pd.DataFrame({
    "Generation": range(1, len(fitness_history) + 1),
    "Best_Fitness_RMSE": fitness_history
}).to_csv(fitness_csv, index=False)

# ============================================================
# Save Models
# ============================================================

h5_path = os.path.join(OUTPUT_DIR, PREFIX + "neural_network_model.h5")
keras_path = os.path.join(OUTPUT_DIR, PREFIX + "neural_network_model.keras")
rf_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "random_forest_model.pkl")
scaler_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
cluster_scaler_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "cluster_scaler.pkl")
encoder_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl")
kmeans_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "kmeans_model.pkl")
feature_mask_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "feature_mask.pkl")

csa_nn_model.save(h5_path)
csa_nn_model.save(keras_path)

joblib.dump(csa_rf_model, rf_pkl_path)
joblib.dump(scaler, scaler_pkl_path)
joblib.dump(cluster_scaler, cluster_scaler_pkl_path)
joblib.dump(label_encoders, encoder_pkl_path)
joblib.dump(kmeans, kmeans_pkl_path)
joblib.dump(best_mask, feature_mask_pkl_path)

# ============================================================
# Graph Paths
# ============================================================

fitness_graph = os.path.join(OUTPUT_DIR, PREFIX + "fitness_graph.png")
accuracy_graph = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")
heatmap_graph = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")
comparison_graph = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")
result_graph = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")
prediction_graph = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")
actual_vs_predicted_graph = os.path.join(OUTPUT_DIR, PREFIX + "actual_vs_predicted_graph.png")
nn_loss_graph = os.path.join(OUTPUT_DIR, PREFIX + "nn_loss_graph.png")
cluster_graph = os.path.join(OUTPUT_DIR, PREFIX + "cluster_graph.png")
selected_features_graph = os.path.join(OUTPUT_DIR, PREFIX + "selected_features_graph.png")

# ============================================================
# Graph 1: CSA Fitness Graph
# ============================================================

plt.figure(figsize=(8, 6))
plt.plot(range(1, len(fitness_history) + 1), fitness_history, marker="o")
plt.title("CSA Optimization Fitness Graph")
plt.xlabel("Generation")
plt.ylabel("Best Fitness RMSE")
plt.tight_layout()
plt.savefig(fitness_graph, dpi=300)
plt.show()
plt.close()

# ============================================================
# Graph 2: Accuracy Graph
# ============================================================

plt.figure(figsize=(8, 6))
sns.barplot(data=results_df, x="Model", y="Accuracy_Percentage")
plt.title("CSA Model Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy Percentage")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(accuracy_graph, dpi=300)
plt.show()
plt.close()

# ============================================================
# Graph 3: Heatmap
# ============================================================

plt.figure(figsize=(12, 8))

corr = df[
    [
        "Demand_Tons",
        "Supply_Tons",
        "Storage_Capacity_Tons",
        "Transport_Cost_Per_KM",
        "Road_Quality_Index",
        "Seasonal_Risk_Index",
        "Demand_Supply_Gap",
        "Market_Load_Index",
        "Estimated_Logistics_Cost",
        "CSA_Predicted_Logistics_Cost"
    ]
].corr()

sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("CSA Feature Correlation Heatmap")
plt.tight_layout()
plt.savefig(heatmap_graph, dpi=300)
plt.show()
plt.close()

# ============================================================
# Graph 4: Model Comparison Graph
# ============================================================

comparison_df = results_df.melt(
    id_vars="Model",
    value_vars=["MAE", "RMSE"],
    var_name="Metric",
    value_name="Value"
)

plt.figure(figsize=(9, 6))
sns.barplot(data=comparison_df, x="Model", y="Value", hue="Metric")
plt.title("CSA Model Error Comparison Graph")
plt.xlabel("Model")
plt.ylabel("Error Value")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(comparison_graph, dpi=300)
plt.show()
plt.close()

# ============================================================
# Graph 5: Result Graph
# ============================================================

plt.figure(figsize=(8, 6))
sns.barplot(data=results_df, x="Model", y="R2_Score")
plt.title("CSA R2 Score Result Graph")
plt.xlabel("Model")
plt.ylabel("R2 Score")
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(result_graph, dpi=300)
plt.show()
plt.close()

# ============================================================
# Graph 6: Prediction Graph
# ============================================================

top_pred_df = df.sort_values(
    by="CSA_Predicted_Logistics_Cost",
    ascending=False
).head(20)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_pred_df,
    x=market_col,
    y="CSA_Predicted_Logistics_Cost"
)
plt.title("CSA Top 20 Markets by Predicted Logistics Cost")
plt.xlabel("Market")
plt.ylabel("Predicted Logistics Cost")
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig(prediction_graph, dpi=300)
plt.show()
plt.close()

# ============================================================
# Graph 7: Actual vs Predicted Graph
# ============================================================

plt.figure(figsize=(8, 6))
plt.scatter(
    df["Estimated_Logistics_Cost"],
    df["CSA_Predicted_Logistics_Cost"]
)
plt.xlabel("Actual Logistics Cost")
plt.ylabel("CSA Predicted Logistics Cost")
plt.title("CSA Actual vs Predicted Logistics Cost")
plt.tight_layout()
plt.savefig(actual_vs_predicted_graph, dpi=300)
plt.show()
plt.close()

# ============================================================
# Graph 8: Neural Network Loss Graph
# ============================================================

plt.figure(figsize=(8, 6))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("CSA Neural Network Training Loss Graph")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(nn_loss_graph, dpi=300)
plt.show()
plt.close()

# ============================================================
# Graph 9: Cluster Graph
# ============================================================

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="Demand_Tons",
    y="Supply_Tons",
    hue="CSA_Cluster_Label"
)
plt.title("CSA Supply Chain Market Cluster Graph")
plt.xlabel("Demand Tons")
plt.ylabel("Supply Tons")
plt.tight_layout()
plt.savefig(cluster_graph, dpi=300)
plt.show()
plt.close()

# ============================================================
# Graph 10: Selected Features Graph
# ============================================================

selected_feature_df = pd.DataFrame({
    "Feature": feature_cols,
    "Selected": best_mask
})

plt.figure(figsize=(10, 6))
sns.barplot(
    data=selected_feature_df,
    x="Feature",
    y="Selected"
)
plt.title("CSA Selected Feature Graph")
plt.xlabel("Feature")
plt.ylabel("Selected Status")
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig(selected_features_graph, dpi=300)
plt.show()
plt.close()

# ============================================================
# YAML Config
# ============================================================

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

yaml_data = {
    "project_name": "CSA Smart Agricultural Supply Chain Optimization System",
    "author": "Sagnik Patra",
    "algorithm": "Cuckoo Search Algorithm",
    "input_file": INPUT_FILE,
    "output_directory": OUTPUT_DIR,
    "prefix": PREFIX,
    "target_column": target_col,
    "all_features": feature_cols,
    "selected_features": selected_features,
    "best_model": str(best_model_name),
    "csa_parameters": {
        "n_nests": 20,
        "generations": 30,
        "discovery_rate": 0.25,
        "best_fitness": float(best_fitness)
    },
    "generated_files": {
        "processed_csv": processed_csv,
        "result_csv": result_csv,
        "prediction_csv": prediction_csv,
        "selected_features_csv": selected_features_csv,
        "cluster_csv": cluster_csv,
        "fitness_csv": fitness_csv,
        "h5_model": h5_path,
        "keras_model": keras_path,
        "random_forest_pkl": rf_pkl_path,
        "scaler_pkl": scaler_pkl_path,
        "cluster_scaler_pkl": cluster_scaler_pkl_path,
        "encoder_pkl": encoder_pkl_path,
        "kmeans_pkl": kmeans_pkl_path,
        "feature_mask_pkl": feature_mask_pkl_path,
        "fitness_graph": fitness_graph,
        "accuracy_graph": accuracy_graph,
        "heatmap_graph": heatmap_graph,
        "comparison_graph": comparison_graph,
        "result_graph": result_graph,
        "prediction_graph": prediction_graph,
        "actual_vs_predicted_graph": actual_vs_predicted_graph,
        "nn_loss_graph": nn_loss_graph,
        "cluster_graph": cluster_graph,
        "selected_features_graph": selected_features_graph
    }
}

with open(yaml_path, "w") as f:
    yaml.dump(yaml_data, f, sort_keys=False)

# ============================================================
# JSON Summary
# ============================================================

json_path = os.path.join(OUTPUT_DIR, PREFIX + "summary.json")

json_data = {
    "project_name": "CSA Smart Agricultural Supply Chain Optimization System",
    "author": "Sagnik Patra",
    "algorithm": "Cuckoo Search Algorithm",
    "dataset_shape": list(df.shape),
    "total_markets": int(len(df)),
    "market_column": market_col,
    "district_column": district_col,
    "best_model": str(best_model_name),
    "selected_features": selected_features,
    "best_fitness": float(best_fitness),
    "model_results": results_df.to_dict(orient="records"),
    "files_saved": yaml_data["generated_files"]
}

with open(json_path, "w") as f:
    json.dump(json_data, f, indent=4)

# ============================================================
# Final Output
# ============================================================

print("\n===================================================")
print("CSA PROJECT COMPLETED SUCCESSFULLY")
print("===================================================")

print("\nCSV Files:")
print(processed_csv)
print(result_csv)
print(prediction_csv)
print(selected_features_csv)
print(cluster_csv)
print(fitness_csv)

print("\nGraphs:")
print(fitness_graph)
print(accuracy_graph)
print(heatmap_graph)
print(comparison_graph)
print(result_graph)
print(prediction_graph)
print(actual_vs_predicted_graph)
print(nn_loss_graph)
print(cluster_graph)
print(selected_features_graph)

print("\nModels:")
print(h5_path)
print(keras_path)
print(rf_pkl_path)
print(scaler_pkl_path)
print(cluster_scaler_pkl_path)
print(encoder_pkl_path)
print(kmeans_pkl_path)
print(feature_mask_pkl_path)

print("\nConfig Files:")
print(yaml_path)
print(json_path)

print("\nBest Model:", best_model_name)

print("\nCSA Selected Features:")
for feature in selected_features:
    print("-", feature)

print("\nModel Results:")
print(results_df)
```
