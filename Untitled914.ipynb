```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.cluster import KMeans

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ============================================================
# Paths
# ============================================================

INPUT_FILE = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Supply Chain Optimization System\megambmarketname_n.xls"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Supply Chain Optimization System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "smart_agri_supply_chain_"

# ============================================================
# Load Dataset
# ============================================================

df = pd.read_excel(INPUT_FILE)

df.columns = (
    df.columns.astype(str)
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

df = df.drop_duplicates()
df = df.fillna("Unknown")

print("Dataset loaded successfully")
print("Shape:", df.shape)
print("Columns:", list(df.columns))

# ============================================================
# Detect Market and District Columns
# ============================================================

market_col = None
district_col = None

for col in df.columns:
    if "market" in col.lower():
        market_col = col
    if "district" in col.lower():
        district_col = col

if market_col is None:
    market_col = df.columns[0]

if district_col is None:
    if len(df.columns) > 1:
        district_col = df.columns[1]
    else:
        df["District"] = "Unknown"
        district_col = "District"

print("Market Column:", market_col)
print("District Column:", district_col)

# ============================================================
# Synthetic Supply Chain Features
# ============================================================

np.random.seed(42)

df["Market_ID"] = range(1, len(df) + 1)
df["Latitude"] = np.random.uniform(25.0, 26.8, len(df))
df["Longitude"] = np.random.uniform(89.5, 92.8, len(df))

df["Demand_Tons"] = np.random.randint(50, 500, len(df))
df["Supply_Tons"] = np.random.randint(40, 550, len(df))
df["Storage_Capacity_Tons"] = np.random.randint(100, 1000, len(df))
df["Transport_Cost_Per_KM"] = np.random.uniform(15, 60, len(df))
df["Road_Quality_Index"] = np.random.uniform(0.4, 1.0, len(df))
df["Seasonal_Risk_Index"] = np.random.uniform(0.1, 0.9, len(df))

df["Demand_Supply_Gap"] = df["Demand_Tons"] - df["Supply_Tons"]
df["Market_Load_Index"] = df["Demand_Tons"] / (df["Storage_Capacity_Tons"] + 1)

df["Estimated_Logistics_Cost"] = (
    abs(df["Demand_Supply_Gap"]) * df["Transport_Cost_Per_KM"]
    + df["Seasonal_Risk_Index"] * 1000
    - df["Road_Quality_Index"] * 300
)

df["Estimated_Logistics_Cost"] = df["Estimated_Logistics_Cost"].clip(lower=100)

# ============================================================
# Encoding
# ============================================================

label_encoders = {}

for col in [market_col, district_col]:
    le = LabelEncoder()
    df[col + "_Encoded"] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

# ============================================================
# Features and Target
# ============================================================

feature_cols = [
    market_col + "_Encoded",
    district_col + "_Encoded",
    "Latitude",
    "Longitude",
    "Demand_Tons",
    "Supply_Tons",
    "Storage_Capacity_Tons",
    "Transport_Cost_Per_KM",
    "Road_Quality_Index",
    "Seasonal_Risk_Index",
    "Demand_Supply_Gap",
    "Market_Load_Index"
]

target_col = "Estimated_Logistics_Cost"

X = df[feature_cols]
y = df[target_col]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================================
# ML Models
# ============================================================

models = {
    "Random Forest": RandomForestRegressor(n_estimators=200, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(random_state=42),
    "Decision Tree": DecisionTreeRegressor(random_state=42)
}

results = []

trained_models = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, pred)
    rmse = np.sqrt(mean_squared_error(y_test, pred))
    r2 = r2_score(y_test, pred)

    accuracy_score = max(0, r2) * 100

    results.append({
        "Model": name,
        "MAE": mae,
        "RMSE": rmse,
        "R2_Score": r2,
        "Accuracy_Percentage": accuracy_score
    })

    trained_models[name] = model

# ============================================================
# Neural Network
# ============================================================

nn_model = Sequential([
    Dense(128, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1)
])

nn_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = nn_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

nn_pred = nn_model.predict(X_test).flatten()

nn_mae = mean_absolute_error(y_test, nn_pred)
nn_rmse = np.sqrt(mean_squared_error(y_test, nn_pred))
nn_r2 = r2_score(y_test, nn_pred)
nn_accuracy = max(0, nn_r2) * 100

results.append({
    "Model": "Neural Network",
    "MAE": nn_mae,
    "RMSE": nn_rmse,
    "R2_Score": nn_r2,
    "Accuracy_Percentage": nn_accuracy
})

results_df = pd.DataFrame(results)

# ============================================================
# Best Model
# ============================================================

best_model_name = results_df.sort_values(
    by="R2_Score",
    ascending=False
).iloc[0]["Model"]

print("Best Model:", best_model_name)

if best_model_name == "Neural Network":
    final_pred = nn_model.predict(X_scaled).flatten()
else:
    final_pred = trained_models[best_model_name].predict(X_scaled)

df["Predicted_Logistics_Cost"] = final_pred
df["Market_Rank"] = df["Predicted_Logistics_Cost"].rank(
    ascending=False,
    method="dense"
).astype(int)

# ============================================================
# KMeans Clustering
# ============================================================

cluster_features = df[
    [
        "Demand_Tons",
        "Supply_Tons",
        "Storage_Capacity_Tons",
        "Transport_Cost_Per_KM",
        "Road_Quality_Index",
        "Seasonal_Risk_Index"
    ]
]

cluster_scaler = StandardScaler()
cluster_scaled = cluster_scaler.fit_transform(cluster_features)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Supply_Chain_Cluster"] = kmeans.fit_predict(cluster_scaled)

cluster_labels = {
    0: "Low Priority Market",
    1: "Medium Priority Market",
    2: "High Priority Market"
}

df["Cluster_Label"] = df["Supply_Chain_Cluster"].map(cluster_labels)

# ============================================================
# Save CSV Files
# ============================================================

result_csv = os.path.join(OUTPUT_DIR, PREFIX + "model_results.csv")
prediction_csv = os.path.join(OUTPUT_DIR, PREFIX + "prediction_results.csv")
processed_csv = os.path.join(OUTPUT_DIR, PREFIX + "processed_dataset.csv")
cluster_csv = os.path.join(OUTPUT_DIR, PREFIX + "cluster_results.csv")

results_df.to_csv(result_csv, index=False)

df.to_csv(processed_csv, index=False)

df[
    [
        market_col,
        district_col,
        "Demand_Tons",
        "Supply_Tons",
        "Predicted_Logistics_Cost",
        "Market_Rank",
        "Cluster_Label"
    ]
].to_csv(prediction_csv, index=False)

df[
    [
        market_col,
        district_col,
        "Supply_Chain_Cluster",
        "Cluster_Label"
    ]
].to_csv(cluster_csv, index=False)

# ============================================================
# Save Models
# ============================================================

h5_path = os.path.join(OUTPUT_DIR, PREFIX + "neural_network_model.h5")
rf_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "random_forest_model.pkl")
gb_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "gradient_boosting_model.pkl")
dt_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "decision_tree_model.pkl")
scaler_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
encoder_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl")
kmeans_pkl_path = os.path.join(OUTPUT_DIR, PREFIX + "kmeans_model.pkl")

nn_model.save(h5_path)
joblib.dump(trained_models["Random Forest"], rf_pkl_path)
joblib.dump(trained_models["Gradient Boosting"], gb_pkl_path)
joblib.dump(trained_models["Decision Tree"], dt_pkl_path)
joblib.dump(scaler, scaler_pkl_path)
joblib.dump(label_encoders, encoder_pkl_path)
joblib.dump(kmeans, kmeans_pkl_path)

# ============================================================
# Graph 1: Accuracy Graph
# ============================================================

plt.figure(figsize=(9, 6))
sns.barplot(
    data=results_df,
    x="Model",
    y="Accuracy_Percentage"
)
plt.title("Model Accuracy Comparison")
plt.xlabel("Model")
plt.ylabel("Accuracy Percentage")
plt.xticks(rotation=20)
plt.tight_layout()

accuracy_graph = os.path.join(OUTPUT_DIR, PREFIX + "accuracy_graph.png")
plt.savefig(accuracy_graph, dpi=300)
plt.show()

# ============================================================
# Graph 2: Heatmap
# ============================================================

plt.figure(figsize=(12, 8))
corr = df[
    [
        "Demand_Tons",
        "Supply_Tons",
        "Storage_Capacity_Tons",
        "Transport_Cost_Per_KM",
        "Road_Quality_Index",
        "Seasonal_Risk_Index",
        "Demand_Supply_Gap",
        "Market_Load_Index",
        "Estimated_Logistics_Cost",
        "Predicted_Logistics_Cost"
    ]
].corr()

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Feature Correlation Heatmap")
plt.tight_layout()

heatmap_graph = os.path.join(OUTPUT_DIR, PREFIX + "heatmap.png")
plt.savefig(heatmap_graph, dpi=300)
plt.show()

# ============================================================
# Graph 3: Comparison Graph MAE/RMSE
# ============================================================

comparison_df = results_df.melt(
    id_vars="Model",
    value_vars=["MAE", "RMSE"],
    var_name="Metric",
    value_name="Value"
)

plt.figure(figsize=(10, 6))
sns.barplot(
    data=comparison_df,
    x="Model",
    y="Value",
    hue="Metric"
)
plt.title("Model Error Comparison")
plt.xlabel("Model")
plt.ylabel("Error Value")
plt.xticks(rotation=20)
plt.tight_layout()

comparison_graph = os.path.join(OUTPUT_DIR, PREFIX + "comparison_graph.png")
plt.savefig(comparison_graph, dpi=300)
plt.show()

# ============================================================
# Graph 4: Result Graph - R2 Score
# ============================================================

plt.figure(figsize=(9, 6))
sns.barplot(
    data=results_df,
    x="Model",
    y="R2_Score"
)
plt.title("Model R2 Score Result Graph")
plt.xlabel("Model")
plt.ylabel("R2 Score")
plt.xticks(rotation=20)
plt.tight_layout()

result_graph = os.path.join(OUTPUT_DIR, PREFIX + "result_graph.png")
plt.savefig(result_graph, dpi=300)
plt.show()

# ============================================================
# Graph 5: Prediction Graph
# ============================================================

top_pred_df = df.sort_values(
    by="Predicted_Logistics_Cost",
    ascending=False
).head(20)

plt.figure(figsize=(12, 6))
sns.barplot(
    data=top_pred_df,
    x=market_col,
    y="Predicted_Logistics_Cost"
)
plt.title("Top 20 Markets by Predicted Logistics Cost")
plt.xlabel("Market")
plt.ylabel("Predicted Logistics Cost")
plt.xticks(rotation=90)
plt.tight_layout()

prediction_graph = os.path.join(OUTPUT_DIR, PREFIX + "prediction_graph.png")
plt.savefig(prediction_graph, dpi=300)
plt.show()

# ============================================================
# Graph 6: Actual vs Predicted
# ============================================================

plt.figure(figsize=(8, 6))
plt.scatter(
    df["Estimated_Logistics_Cost"],
    df["Predicted_Logistics_Cost"]
)
plt.xlabel("Actual Logistics Cost")
plt.ylabel("Predicted Logistics Cost")
plt.title("Actual vs Predicted Logistics Cost")
plt.tight_layout()

actual_vs_pred_graph = os.path.join(OUTPUT_DIR, PREFIX + "actual_vs_predicted_graph.png")
plt.savefig(actual_vs_pred_graph, dpi=300)
plt.show()

# ============================================================
# Graph 7: Neural Network Loss Graph
# ============================================================

plt.figure(figsize=(8, 6))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Neural Network Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()

nn_loss_graph = os.path.join(OUTPUT_DIR, PREFIX + "nn_loss_graph.png")
plt.savefig(nn_loss_graph, dpi=300)
plt.show()

# ============================================================
# Graph 8: Cluster Graph
# ============================================================

plt.figure(figsize=(8, 6))
sns.scatterplot(
    data=df,
    x="Demand_Tons",
    y="Supply_Tons",
    hue="Cluster_Label"
)
plt.title("Supply Chain Market Cluster Graph")
plt.xlabel("Demand Tons")
plt.ylabel("Supply Tons")
plt.tight_layout()

cluster_graph = os.path.join(OUTPUT_DIR, PREFIX + "cluster_graph.png")
plt.savefig(cluster_graph, dpi=300)
plt.show()

# ============================================================
# YAML Config
# ============================================================

yaml_data = {
    "project_name": "Smart Agricultural Supply Chain Optimization System",
    "author": "Sagnik Patra",
    "input_file": INPUT_FILE,
    "output_directory": OUTPUT_DIR,
    "target_column": target_col,
    "features_used": feature_cols,
    "models_used": [
        "Random Forest Regressor",
        "Gradient Boosting Regressor",
        "Decision Tree Regressor",
        "Neural Network",
        "KMeans Clustering"
    ],
    "best_model": str(best_model_name),
    "generated_files": {
        "result_csv": result_csv,
        "prediction_csv": prediction_csv,
        "processed_csv": processed_csv,
        "cluster_csv": cluster_csv,
        "accuracy_graph": accuracy_graph,
        "heatmap": heatmap_graph,
        "comparison_graph": comparison_graph,
        "result_graph": result_graph,
        "prediction_graph": prediction_graph,
        "actual_vs_predicted_graph": actual_vs_pred_graph,
        "nn_loss_graph": nn_loss_graph,
        "cluster_graph": cluster_graph,
        "h5_model": h5_path,
        "random_forest_pkl": rf_pkl_path,
        "gradient_boosting_pkl": gb_pkl_path,
        "decision_tree_pkl": dt_pkl_path,
        "scaler_pkl": scaler_pkl_path,
        "encoder_pkl": encoder_pkl_path,
        "kmeans_pkl": kmeans_pkl_path
    }
}

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(yaml_data, f, sort_keys=False)

# ============================================================
# JSON Summary
# ============================================================

json_data = {
    "project_name": "Smart Agricultural Supply Chain Optimization System",
    "author": "Sagnik Patra",
    "dataset_shape": list(df.shape),
    "total_markets": int(len(df)),
    "market_column": market_col,
    "district_column": district_col,
    "best_model": str(best_model_name),
    "model_results": results_df.to_dict(orient="records"),
    "files_saved": yaml_data["generated_files"]
}

json_path = os.path.join(OUTPUT_DIR, PREFIX + "summary.json")

with open(json_path, "w") as f:
    json.dump(json_data, f, indent=4)

# ============================================================
# Final Output
# ============================================================

print("\n===================================================")
print("PROJECT COMPLETED SUCCESSFULLY")
print("===================================================")

print("\nSaved CSV Files:")
print(result_csv)
print(prediction_csv)
print(processed_csv)
print(cluster_csv)

print("\nSaved Graphs:")
print(accuracy_graph)
print(heatmap_graph)
print(comparison_graph)
print(result_graph)
print(prediction_graph)
print(actual_vs_pred_graph)
print(nn_loss_graph)
print(cluster_graph)

print("\nSaved Models:")
print(h5_path)
print(rf_pkl_path)
print(gb_pkl_path)
print(dt_pkl_path)

print("\nSaved Config Files:")
print(yaml_path)
print(json_path)

print("\nBest Model:", best_model_name)
print("\nModel Results:")
print(results_df)
```
