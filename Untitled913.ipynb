```python
import os
import json
import yaml
import joblib
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.cluster import KMeans
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

warnings.filterwarnings("ignore")

# ============================================================
# Paths
# ============================================================

INPUT_FILE = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Supply Chain Optimization System\megambmarketname_n.xls"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Agricultural Supply Chain Optimization System"

os.makedirs(OUTPUT_DIR, exist_ok=True)

PREFIX = "smart_agri_supply_chain_"

# ============================================================
# Load Dataset
# ============================================================

df = pd.read_excel(INPUT_FILE)

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("Columns:", list(df.columns))
print(df.head())

# ============================================================
# Clean Column Names
# ============================================================

df.columns = (
    df.columns
    .astype(str)
    .str.strip()
    .str.replace(" ", "_")
    .str.replace("-", "_")
)

# ============================================================
# Basic Cleaning
# ============================================================

df = df.drop_duplicates()
df = df.fillna("Unknown")

# Try to detect useful columns
market_col = None
district_col = None

for col in df.columns:
    low = col.lower()
    if "market" in low:
        market_col = col
    if "district" in low:
        district_col = col

if market_col is None:
    market_col = df.columns[0]

if district_col is None:
    if len(df.columns) > 1:
        district_col = df.columns[1]
    else:
        df["District"] = "Unknown"
        district_col = "District"

print("Market Column:", market_col)
print("District Column:", district_col)

# ============================================================
# Synthetic Supply Chain Feature Generation
# Because this file mainly contains market names/district names
# ============================================================

np.random.seed(42)

df["Market_ID"] = range(1, len(df) + 1)

df["Latitude"] = np.random.uniform(25.0, 26.8, len(df))
df["Longitude"] = np.random.uniform(89.5, 92.8, len(df))

df["Demand_Tons"] = np.random.randint(50, 500, len(df))
df["Supply_Tons"] = np.random.randint(40, 550, len(df))
df["Storage_Capacity_Tons"] = np.random.randint(100, 1000, len(df))
df["Transport_Cost_Per_KM"] = np.random.uniform(15, 60, len(df))
df["Road_Quality_Index"] = np.random.uniform(0.4, 1.0, len(df))
df["Seasonal_Risk_Index"] = np.random.uniform(0.1, 0.9, len(df))

df["Demand_Supply_Gap"] = df["Demand_Tons"] - df["Supply_Tons"]
df["Market_Load_Index"] = df["Demand_Tons"] / (df["Storage_Capacity_Tons"] + 1)

# Target variable: estimated logistics cost
df["Estimated_Logistics_Cost"] = (
    abs(df["Demand_Supply_Gap"]) * df["Transport_Cost_Per_KM"]
    + df["Seasonal_Risk_Index"] * 1000
    - df["Road_Quality_Index"] * 300
)

df["Estimated_Logistics_Cost"] = df["Estimated_Logistics_Cost"].clip(lower=100)

# ============================================================
# Label Encoding
# ============================================================

label_encoders = {}

for col in [market_col, district_col]:
    le = LabelEncoder()
    df[col + "_Encoded"] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

# ============================================================
# Feature Selection
# ============================================================

feature_cols = [
    market_col + "_Encoded",
    district_col + "_Encoded",
    "Latitude",
    "Longitude",
    "Demand_Tons",
    "Supply_Tons",
    "Storage_Capacity_Tons",
    "Transport_Cost_Per_KM",
    "Road_Quality_Index",
    "Seasonal_Risk_Index",
    "Demand_Supply_Gap",
    "Market_Load_Index"
]

target_col = "Estimated_Logistics_Cost"

X = df[feature_cols]
y = df[target_col]

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ============================================================
# Train Test Split
# ============================================================

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

# ============================================================
# Random Forest Model
# ============================================================

rf_model = RandomForestRegressor(
    n_estimators=200,
    random_state=42,
    max_depth=10
)

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)

rf_mae = mean_absolute_error(y_test, rf_pred)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_r2 = r2_score(y_test, rf_pred)

print("\nRandom Forest Results")
print("MAE:", rf_mae)
print("RMSE:", rf_rmse)
print("R2 Score:", rf_r2)

# ============================================================
# Neural Network Model
# ============================================================

nn_model = Sequential([
    Dense(128, activation="relu", input_shape=(X_train.shape[1],)),
    Dropout(0.2),
    Dense(64, activation="relu"),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dense(1)
])

nn_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

early_stop = EarlyStopping(
    monitor="val_loss",
    patience=20,
    restore_best_weights=True
)

history = nn_model.fit(
    X_train,
    y_train,
    validation_split=0.2,
    epochs=200,
    batch_size=8,
    callbacks=[early_stop],
    verbose=1
)

nn_pred = nn_model.predict(X_test).flatten()

nn_mae = mean_absolute_error(y_test, nn_pred)
nn_rmse = np.sqrt(mean_squared_error(y_test, nn_pred))
nn_r2 = r2_score(y_test, nn_pred)

print("\nNeural Network Results")
print("MAE:", nn_mae)
print("RMSE:", nn_rmse)
print("R2 Score:", nn_r2)

# ============================================================
# KMeans Market Clustering
# ============================================================

cluster_features = df[
    [
        "Demand_Tons",
        "Supply_Tons",
        "Storage_Capacity_Tons",
        "Transport_Cost_Per_KM",
        "Road_Quality_Index",
        "Seasonal_Risk_Index"
    ]
]

cluster_scaled = scaler.fit_transform(cluster_features)

kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Supply_Chain_Cluster"] = kmeans.fit_predict(cluster_scaled)

cluster_map = {
    0: "Low Priority Market",
    1: "Medium Priority Market",
    2: "High Priority Market"
}

df["Cluster_Label"] = df["Supply_Chain_Cluster"].map(cluster_map)

# ============================================================
# Haversine Distance Function
# ============================================================

def haversine(lat1, lon1, lat2, lon2):
    R = 6371
    lat1, lon1, lat2, lon2 = map(
        np.radians,
        [lat1, lon1, lat2, lon2]
    )
    dlat = lat2 - lat1
    dlon = lon2 - lon1

    a = (
        np.sin(dlat / 2) ** 2
        + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
    )

    c = 2 * np.arcsin(np.sqrt(a))
    return R * c

# ============================================================
# Simple Route Optimization
# Nearest Neighbor Algorithm
# ============================================================

markets = df[[market_col, district_col, "Latitude", "Longitude"]].copy()
markets = markets.reset_index(drop=True)

visited = [0]
route = [0]
total_distance = 0

while len(visited) < len(markets):
    current = route[-1]
    current_lat = markets.loc[current, "Latitude"]
    current_lon = markets.loc[current, "Longitude"]

    nearest_market = None
    nearest_distance = float("inf")

    for i in range(len(markets)):
        if i not in visited:
            dist = haversine(
                current_lat,
                current_lon,
                markets.loc[i, "Latitude"],
                markets.loc[i, "Longitude"]
            )

            if dist < nearest_distance:
                nearest_distance = dist
                nearest_market = i

    visited.append(nearest_market)
    route.append(nearest_market)
    total_distance += nearest_distance

optimized_route_df = markets.loc[route].copy()
optimized_route_df["Route_Order"] = range(1, len(optimized_route_df) + 1)

optimized_route_df["Next_Market"] = optimized_route_df[market_col].shift(-1)
optimized_route_df["Next_Market"] = optimized_route_df["Next_Market"].fillna("End")

# ============================================================
# Predictions on Full Dataset
# ============================================================

X_full_scaled = StandardScaler().fit_transform(X)

df["RF_Predicted_Logistics_Cost"] = rf_model.predict(X_scaled)
df["NN_Predicted_Logistics_Cost"] = nn_model.predict(X_scaled).flatten()

df["Final_Predicted_Logistics_Cost"] = (
    df["RF_Predicted_Logistics_Cost"] + df["NN_Predicted_Logistics_Cost"]
) / 2

df["Market_Rank"] = df["Final_Predicted_Logistics_Cost"].rank(
    ascending=False,
    method="dense"
).astype(int)

# ============================================================
# Save CSV Outputs
# ============================================================

processed_csv = os.path.join(OUTPUT_DIR, PREFIX + "processed_dataset.csv")
prediction_csv = os.path.join(OUTPUT_DIR, PREFIX + "predictions.csv")
route_csv = os.path.join(OUTPUT_DIR, PREFIX + "optimized_route.csv")
cluster_csv = os.path.join(OUTPUT_DIR, PREFIX + "market_clusters.csv")

df.to_csv(processed_csv, index=False)
df[[market_col, district_col, "Final_Predicted_Logistics_Cost", "Market_Rank"]].to_csv(
    prediction_csv,
    index=False
)
optimized_route_df.to_csv(route_csv, index=False)
df[[market_col, district_col, "Supply_Chain_Cluster", "Cluster_Label"]].to_csv(
    cluster_csv,
    index=False
)

# ============================================================
# Save Models
# ============================================================

h5_path = os.path.join(OUTPUT_DIR, PREFIX + "neural_network_model.h5")
pkl_model_path = os.path.join(OUTPUT_DIR, PREFIX + "random_forest_model.pkl")
pkl_scaler_path = os.path.join(OUTPUT_DIR, PREFIX + "scaler.pkl")
pkl_kmeans_path = os.path.join(OUTPUT_DIR, PREFIX + "kmeans_model.pkl")
pkl_encoder_path = os.path.join(OUTPUT_DIR, PREFIX + "label_encoders.pkl")

nn_model.save(h5_path)
joblib.dump(rf_model, pkl_model_path)
joblib.dump(scaler, pkl_scaler_path)
joblib.dump(kmeans, pkl_kmeans_path)
joblib.dump(label_encoders, pkl_encoder_path)

# ============================================================
# Save YAML Configuration
# ============================================================

config = {
    "project_name": "Smart Agricultural Supply Chain Optimization System",
    "author": "Sagnik Patra",
    "input_file": INPUT_FILE,
    "output_directory": OUTPUT_DIR,
    "target_column": target_col,
    "features_used": feature_cols,
    "models": {
        "machine_learning": "RandomForestRegressor",
        "deep_learning": "Neural Network",
        "clustering": "KMeans",
        "route_optimization": "Nearest Neighbor Algorithm"
    },
    "outputs": {
        "h5_model": h5_path,
        "pkl_model": pkl_model_path,
        "yaml_config": PREFIX + "config.yaml",
        "json_summary": PREFIX + "summary.json"
    }
}

yaml_path = os.path.join(OUTPUT_DIR, PREFIX + "config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(config, f, sort_keys=False)

# ============================================================
# Save JSON Summary
# ============================================================

summary = {
    "dataset_shape": df.shape,
    "market_column": market_col,
    "district_column": district_col,
    "total_markets": int(len(df)),
    "total_optimized_route_distance_km": float(total_distance),
    "random_forest_results": {
        "mae": float(rf_mae),
        "rmse": float(rf_rmse),
        "r2_score": float(rf_r2)
    },
    "neural_network_results": {
        "mae": float(nn_mae),
        "rmse": float(nn_rmse),
        "r2_score": float(nn_r2)
    },
    "generated_files": [
        processed_csv,
        prediction_csv,
        route_csv,
        cluster_csv,
        h5_path,
        pkl_model_path,
        pkl_scaler_path,
        pkl_kmeans_path,
        pkl_encoder_path,
        yaml_path
    ]
}

json_path = os.path.join(OUTPUT_DIR, PREFIX + "summary.json")

with open(json_path, "w") as f:
    json.dump(summary, f, indent=4)

# ============================================================
# Visualization Graphs
# ============================================================

plt.figure(figsize=(10, 6))
plt.bar(
    df[market_col].astype(str).head(20),
    df["Final_Predicted_Logistics_Cost"].head(20)
)
plt.xticks(rotation=90)
plt.title("Top Market Logistics Cost Prediction")
plt.xlabel("Market")
plt.ylabel("Predicted Logistics Cost")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "logistics_cost_graph.png"))
plt.close()

plt.figure(figsize=(8, 6))
plt.scatter(
    df["Demand_Tons"],
    df["Supply_Tons"],
    c=df["Supply_Chain_Cluster"]
)
plt.title("Market Clustering Based on Demand and Supply")
plt.xlabel("Demand Tons")
plt.ylabel("Supply Tons")
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "market_cluster_graph.png"))
plt.close()

plt.figure(figsize=(8, 6))
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.title("Neural Network Training Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.tight_layout()
plt.savefig(os.path.join(OUTPUT_DIR, PREFIX + "nn_training_loss_graph.png"))
plt.close()

# ============================================================
# Final Output
# ============================================================

print("\n============================================================")
print("Smart Agricultural Supply Chain Optimization Completed")
print("============================================================")
print("All files saved in:")
print(OUTPUT_DIR)
print("\nGenerated Main Files:")
print("H5 Model:", h5_path)
print("PKL Model:", pkl_model_path)
print("YAML Config:", yaml_path)
print("JSON Summary:", json_path)
print("Optimized Route CSV:", route_csv)
print("Prediction CSV:", prediction_csv)
print("============================================================")
```
